````python
from collections import deque
class BinaryTree:
    def __init__(self, value):
        self.__data = value
        self.__right = None
        self.__left = None
        self.__parent = None
        self.__depth = 0

    def getValue(self):
        return self.__data
    def setValue(self, newValue):
        self.__data = newValue

    def getParent(self):
        return self.__parent
    def setParent(self, tree):
        self.__parent = tree

    def getRight(self):
        return self.__right
    def getLeft(self):
        return self.__left

    def insertRight(self, tree):
        if self.__right == None:
            self.__right = tree
            tree.setParent(self)
            tree.setDepth(self.__depth + 1)

    def insertLeft(self, tree):
        if self.__left == None:
            self.__left = tree
            tree.setParent(self)
            tree.setDepth(self.__depth + 1)

    def deleteRight(self):
        self.__right = None

    def deleteLeft(self):
        self.__left = None
     
    def setDepth(self, depth):
        self.__depth = depth   
    def getDepth(self):
        return self.__depth
    

        
    ### Exercise 1: Count Nodes
    def count_nodes(self):
        """Implement a method to count the total number of nodes in the binary tree."""
        count = 0
        if self != None:
            level = deque()
            level.append(self)

            while len(level) > 0:
                #print(len(level))
                cur = level.popleft()
                count += 1
                #print(cur.getValue())
                r = cur.getRight()
                l = cur.getLeft()
                if l != None:
                    #print("from {} Appending: {}".format(cur.getValue(),
                    #                                      l.getValue()))
                    level.append(l)
                if r != None:
                    level.append(r)
        return count
        

            
    ### Exercise 2: Depth of Tree
    def tree_depth(self):
        """Implement a method to find the max depth (height) of the binary tree."""
        max_depth = 0
        if self != None:
            level = deque()
            level.append(self)

            while len(level) > 0:               
                cur = level.popleft()
                if cur.getDepth() > max_depth:
                    max_depth = cur.getDepth()
                r = cur.getRight()
                l = cur.getLeft()
                if l != None:
                    level.append(l)
                if r != None:
                    level.append(r)
                    
        return max_depth

    ### Exercise 3: In-order Traversal
    def in_order_traversal(self):
        """Implement a method to perform in-order traversal of the binary tree and print the values."""
        if self != None:
            r = self.getRight()
            l = self.getLeft()
            print(self.getValue()) # print before recursive calls!
            if l != None:
                l.preOrderDFS()
            #print(self.getValue()) # print before recursive calls!
            if r != None:
                r.preOrderDFS()
    
    ### Exercise 4: Post-order Traversal
    def post_order_traversal(self):
        """Implement a method to perform post-order traversal of the binary tree and print the values."""
        if self != None:
            r = self.getRight()
            l = self.getLeft()
            
            if l != None:
                l.preOrderDFS()
            #print(self.getValue()) # print before recursive calls!
            if r != None:
                r.preOrderDFS()
    
    ### Exercise 5: Pre-order Traversal
    def preOrderDFS(self):
        """Implement a method to perform pre-order traversal of the binary tree and print the values."""
        if self != None:
            r = self.getRight()
            l = self.getLeft()
            #print(self.getValue()) # print before recursive calls!
            if l != None:
                l.preOrderDFS()
            if r != None:
                r.preOrderDFS()
    
    ### Exercise 6: Check if Binary Tree is Full
    def is_full_binary_tree(self):
        """Implement a method to check if the binary tree is a full binary tree.
           A tree is a perfect binary if all interval nodes has
           exactly two children and all leaf nodes are on the same level"""
        
        if self is not None:
            level = deque()
            level.append(self)

            while len(level) > 0:
                cur = level.popleft()
                r = cur.getRight()
                l = cur.getLeft()
                
                
                if (l is None and r is not None) or (l is not None and r is None):
                    return False
            
                if l is not None:
                    level.append(l)
                if r is not None:
                    level.append(r)
                
        return True    
    
    ### Exercise 7: get the width of the treee
    def getWidth(self):
        width = 0
        if self != None:
            level = deque()
            level.append(self)

            while len(level) > 0:
                if width < len(level):
                    width = len(level)
                cur = level.popleft()
                #print(cur.getValue())
                r = cur.getRight()
                l = cur.getLeft()
                if l != None:
                    #print("from {} Appending: {}".format(cur.getValue(),
                    #                                      l.getValue()))
                    level.append(l)
                if r != None:
                    level.append(r)
        return width
    
    def BFS(self):
        if self != None:
            level = deque()
            level.append(self)

            while len(level) > 0:
                #print(len(level))
                cur = level.popleft()
                #print(cur.getValue())
                r = cur.getRight()
                l = cur.getLeft()
                if l != None:
                    #print("from {} Appending: {}".format(cur.getValue(),
                    #                                      l.getValue()))
                    level.append(l)
                if r != None:
                    level.append(r)
                    #print("from {} Appending: {}".format(cur.getValue(),
                    #                                     r.getValue()))

def printTree(root, level=0, prefix="Root"):
    if root is not None:
        print("  " * level + f"{prefix}({root.getValue()})")
        if root.getRight() is not None or root.getLeft() is not None:
            printTree(root.getRight(), level + 1, "R->")
            printTree(root.getLeft(), level + 1, "L->")
            
            
def createBST(inList):
    """gets a list of integers (or strings) and returns a BinaryTree that is a Binary Search Tree"""
    BST = None
    
    if len(inList) > 0:
        BST = BinaryTree(inList[0]) # set the first element as root
        inList = inList.pop(0) # remove the first element
        
        for i in inList:
            print(f"Node_ {i}")
            BinaryTree(i)
            
        printTree(BST)
            
        


# if __name__ == "__main__":
BT = BinaryTree("Root")
bt1 = BinaryTree(1)
bt2 = BinaryTree(2)
bt3 = BinaryTree(3)
bt4 = BinaryTree(4)
bt5 = BinaryTree(5)
bt6 = BinaryTree(6)
bt5a = BinaryTree("5a")
bt5b = BinaryTree("5b")
bt5c = BinaryTree("5c")

BT.insertLeft(bt1)
BT.insertRight(bt2)
bt2.insertLeft(bt3)
bt3.insertLeft(bt4)
bt3.insertRight(bt5)
bt2.insertRight(bt6)
bt1.insertRight(bt5b)
bt1.insertLeft(bt5a)
bt5b.insertRight(bt5c)

print("Number of nodes: ", BT.count_nodes())
print("Tree depth: ", BT.tree_depth())
print("In-order traversal:", end=" ")
BT.in_order_traversal()
print("\nPost-order traversal:", end=" ")
BT.post_order_traversal()
print("\nPre-order traversal:", end=" ")
#BT.pre_order_traversal()
print("\n Is a full binary tree?", BT.is_full_binary_tree())

#print(bt4.getDepth())
print("Width of the tree: ", BT.getWidth() )

inList = [5,6,3,10,4,7,9,11]
print("Binary Search Tree: ", createBST(inList))
````
